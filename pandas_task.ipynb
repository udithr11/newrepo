```python
#13   Total contributation hour for each and every agents weekly basis
```

```python
df1['duration_in_hr']=pd.to_timedelta(df1['Duration'])
```

```python
df1['total_duration_in_hr']=list(map(lambda x:x.total_seconds(),df1['duration_in_hr']))
```

```python
df1['total_duration_in_hr']=(df1['total_duration_in_hr'])/(60*60)
```

```python
df1['week']=pd.to_datetime(df1['Date']).dt.weekofyear
```

```python
df1.groupby(['Agent','week'])['total_duration_in_hr'].sum().reset_index()
```

```python
#14   totalpercentage of active hour for a month
```

```python
df1['month']=pd.to_datetime(df1['Date']).dt.month
```

```python
df11=df1.groupby(['Agent','month'])['total_duration_in_hr'].sum().reset_index()
```

```python
df11['percentage_of_active_hr_for_a_month']=((df11['total_duration_in_hr'])/(11*24))*100
#the given table AgentLogingReport.csv is only from 21st to 30th july. so total hours in 11 days=11*24
#to find the percentage of active hour for the given time= (total_duration_in_hr/(11*24))*100
```
